```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import datetime
```

```python
# Each record describes characteristics of a client
processed_table = pd.read_csv('./data/processed/processed_table.csv')
processed_table = processed_table[processed_table['account_year'] < 1998]
```

```python
processed_table.head()
for column in processed_table.columns[0:]:
    print(column)
```

```python
loan_granted = processed_table[processed_table['loan_granted'] == 1]
loan_not_granted = processed_table[processed_table['loan_granted'] != 1]
```

```python
fig, ax = plt.subplots(1, 2,figsize=(8,4))
ax[0].hist(loan_granted['num_of_transfer_otherbank'], color='Red')
ax[1].hist(loan_not_granted['num_of_transfer_otherbank'], color='Black', alpha=0.7)
plt.show()
```
